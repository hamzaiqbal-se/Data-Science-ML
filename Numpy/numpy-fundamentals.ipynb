### Creating Numpy Arrays


```python
# np.array
import numpy as np

a = np.array([1,2,3])
print (a)
print (type(a))
```

```python
# 2D and 3D
b = np.array([[1,2,3],[4,5,6]])
print (b)

c = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print (c)
```

```python
# dtype

a = np.array([1,2,3], dtype=float)
print (a)
b = np.array([1,2,3], dtype=bool)
print (b)
```

```python
# np.arange

np.arange(1,11)
```

```python
# with reshape

np.arange(1,11).reshape(5,2)
```

```python
# np.ones and np.zeros

a = np.ones((3,4))
print(a)
b = np.zeros((3,4))
print(b)
```

```python
# np.random

np.random.random((3,3))
```

```python
# np.linspace

np.linspace(-10,10,10)
```

```python
# np.identity

np.identity(4)
```

### Array Attributes

```python
a1 = np.arange(10,dtype=np.int32)
a2 = np.arange(12,dtype=float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```

```python
# ndim means number of dimensions

a1.ndim
a2.ndim
a3.ndim
```

```python
# shape  ( how much items exist in every dimension)

a1.shape
a2.shape
a3.shape
```

```python
# size means number od items

a1.size
a2.size
a3.size
```

```python
# itemsize ( every item take how much size i memory)

a1.itemsize
a2.itemsize
a3.itemsize
```

```python
# dtype ( datatype of items)
 
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

### Changing Datatype

```python
# astype

a3.astype(np.int32)
```

### Array Operations

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
```

```python
# scalar operations

# arithmetic

a1 * 2
```

```python
# relational

print(a2 > 15)
print(a2 < 15)
print(a2 == 15)
```

```python
# vector operations
# arithmetic

print(a1 + a2)
```

### Array Functions

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
# max/min/sum/prod
# 0 -> col and 1 -> row

print (np.max(a1))
print (np.min(a1))
print (np.sum(a1))
print (np.prod(a1))
```

```python
# mean/median/std/var

print (np.mean(a1))
print (np.median(a1))
print (np.std(a1))
print (np.var(a1))
```

```python
# trigonomoetric functions

print (np.sin(a1))
print (np.cos(a1))
```

```python
# dot product

a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)

np.dot(a2,a3)
```

```python
# log and exponents

print (np.log(a1))
print (np.exp(a1))
```

```python
# round/floor/ceil

print (np.round(np.random.random((2,3))*100))
print (np.floor(np.random.random((2,3))*100))
print (np.ceil(np.random.random((2,3))*100))
```

### Reshaping

### Stacking

```python
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
print(a4)
print(a5)
```

```python
# Horizontal Stacking
print (np.hstack((a4,a5)))
```

```python
# Vertical Stacking
print (np.vstack((a4,a5)))
```

### Splitting

```python
# Horizontal Splitting

print (np.hsplit(a4,2))
```

```python
# Vertical Splittig
a6 = np.arange(1,17).reshape(4,4)
print(a6)
print (np.vsplit(a6,4))
```
